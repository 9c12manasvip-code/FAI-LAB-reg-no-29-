```python
#exp 15 hill climbing
import random

num_tasks = 200
num_servers = 10

# generate random task sizes
tasks = [random.randint(5, 50) for _ in range(num_tasks)]

def calculate_loads(assign):
    loads = [0] * num_servers
    for i in range(num_tasks):
        loads[assign[i]] += tasks[i]
    return loads

def imbalance(loads):
    return max(loads) - min(loads)

def hill_climbing():

    # random starting assignment
    current = [random.randint(0, num_servers-1) for _ in range(num_tasks)]
    current_score = imbalance(calculate_loads(current))

    while True:
        best_neighbor = None
        best_score = current_score

        for i in range(num_tasks):
            for s in range(num_servers):

                if s != current[i]:
                    new = current[:]
                    new[i] = s

                    new_score = imbalance(calculate_loads(new))

                    if new_score < best_score:
                        best_neighbor = new
                        best_score = new_score

        if best_neighbor is None:
            break

        current = best_neighbor
        current_score = best_score

    return current, current_score


best_assign, best_score = hill_climbing()

print("Final loads:", calculate_loads(best_assign))
print("Final imbalance:", best_score)
```
