```python
models = [
    (5, 15),   # Rule-Based
    (25, 55),  # Location Anomaly
    (45, 70),  # Spending Pattern
    (30, 50),  # Device Fingerprint
    (20, 310)   # Merchant Risk
]

MAX_LATENCY = 100
n = len(models)

dp = [[0] * (MAX_LATENCY + 1) for _ in range(n + 1)]

for i in range(1, n + 1):
    weight, value = models[i - 1]
    for w in range(MAX_LATENCY + 1):
        if weight <= w:
            dp[i][w] = max(dp[i - 1][w],
                           dp[i - 1][w - weight] + value)
        else:
            dp[i][w] = dp[i - 1][w]

print("Maximum Fraud Detection Value:", dp[n][MAX_LATENCY])
```
