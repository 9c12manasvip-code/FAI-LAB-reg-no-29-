```python
from itertools import permutations

dist = [
    [0, 10, 15, 20],
    [10, 0, 35, 25],
    [15, 35, 0, 30],
    [20, 25, 30, 0]
]

cities = range(4)

min_cost = float('inf')
best_path = None

for path in permutations(cities):
    cost = 0

    for i in range(3):
        cost += dist[path[i]][path[i+1]]

    cost += dist[path[3]][path[0]]

    if cost < min_cost:
        min_cost = cost
        best_path = path

print("Minimum Distance:", min_cost)
print("Optimal Route:", best_path)
```
