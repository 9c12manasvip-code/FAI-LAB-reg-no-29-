```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression

docs=["mitochondrial disorder muscle weakness",
      "heart disease treatment methods",
      "rare pediatric genetic disorder study",
      "sports nutrition and fitness"]
labels=[1,0,1,0]

vectorizer=TfidfVectorizer()
X=vectorizer.fit_transform(docs)

model=LogisticRegression()
model.fit(X,labels)

query=["pediatric mitochondrial disease"]
q_vec=vectorizer.transform(query)

pred=model.predict(q_vec)
print("Relevant Paper:", "Yes" if pred[0]==1 else "No")
```
