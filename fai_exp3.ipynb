```python
def peak_sqrt(arr):
    low, high = 0, len(arr)-1

    while low < high:
        mid = (low + high) // 2
        if arr[mid] < arr[mid+1]:
            low = mid + 1
        else:
            high = mid

    peak = arr[low]
    return peak, peak ** 0.5

arr = [1, 3, 7, 12, 9, 4, 2]
print(peak_sqrt(arr))
```
