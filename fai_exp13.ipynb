```python
from collections import deque

def bfs_shortest_path(graph, start, goal):
    visited = set()
    queue = deque([[start]])

    while queue:
        path = queue.popleft()
        node = path[-1]

        if node == goal:
            return path

        if node not in visited:
            visited.add(node)

            for neighbour in graph[node]:
                new_path = list(path)
                new_path.append(neighbour)
                queue.append(new_path)

    return "No Path Found"


graph = {
    "A": ["B", "C"],
    "B": ["D", "E"],
    "C": ["F"],
    "D": [],
    "E": ["F"],
    "F": []
}

print("Shortest path:", bfs_shortest_path(graph, "A", "F"))
```

```python
DFS
```

```python
def dfs_search(graph, start, target, visited=None):
    if visited is None:
        visited = set()

    visited.add(start)

    if start == target:
        return [start]

    for neighbour in graph[start]:
        if neighbour not in visited:
            path = dfs_search(graph, neighbour, target, visited)
            if path:
                return [start] + path

    return None


# File system graph
graph = {
    "Root": ["Documents", "Downloads", "Pictures"],
    "Documents": ["Projects", "Resume.pdf"],
    "Projects": [],
    "Resume.pdf": [],
    "Downloads": [],
    "Pictures": []
}

result = dfs_search(graph, "Root", "Resume.pdf")
print("File Path:", result)
```
