```python
import math

def evaluate(state):
    return state

def decision_tree(depth, index, maximizing, values, alpha, beta):
    if depth == 3:
        return evaluate(values[index])

    if maximizing:
        best = -math.inf
        for i in range(2):
            val = decision_tree(depth+1, index*2+i, False, values, alpha, beta)
            best = max(best, val)
            alpha = max(alpha, best)
            if beta <= alpha:
                break
        return best
    else:
        best = math.inf
        for i in range(2):
            val = decision_tree(depth+1, index*2+i, True, values, alpha, beta)
            best = min(best, val)
            beta = min(beta, best)
            if beta <= alpha:
                break
        return best

values = [8, 5, 6, 9, 1, 2, 0, 3]
optimal = decision_tree(0, 0, True, values, -math.inf, math.inf)
print("Optimal Cloud Resource Allocation Score:", optimal)
```
