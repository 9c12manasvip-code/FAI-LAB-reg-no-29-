```python
def check_win(board, player):
    win_conditions = [
        [0, 1, 2], [3, 4, 5], [6, 7, 8], # rows
        [0, 3, 6], [1, 4, 7], [2, 5, 8], # columns
        [0, 4, 8], [2, 4, 6]            # diagonals
    ]
    for condition in win_conditions:
        if all(board[i] == player for i in condition):
            return True
    return False

def is_full(board):
    return ' ' not in board

def evaluate(board):
    if check_win(board, 'X'):
        return 1
    elif check_win(board, 'O'):
        return -1
    elif is_full(board):
        return 0
    else:
        return 0 # Game is ongoing, or can be a draw eventually

def minimax(board, depth, is_maximizing_player):
    score = evaluate(board)

    if score == 1: # X wins
        return score - depth
    if score == -1: # O wins
        return score + depth
    if is_full(board): # Draw
        return 0

    if is_maximizing_player:
        best_score = -float('inf')
        for i in range(9):
            if board[i] == ' ':
                board[i] = 'X'
                best_score = max(best_score, minimax(board, depth + 1, False))
                board[i] = ' ' # Undo the move
        return best_score
    else:
        best_score = float('inf')
        for i in range(9):
            if board[i] == ' ':
                board[i] = 'O'
                best_score = min(best_score, minimax(board, depth + 1, True))
                board[i] = ' ' # Undo the move
        return best_score

def best_decision(board):
    best_score = -float('inf')
    best_move = -1
    for i in range(9):
        if board[i] == ' ':
            board[i] = 'X'
            score = minimax(board, 0, False)
            board[i] = ' ' # Undo the move
            if score > best_score:
                best_score = score
                best_move = i
    return best_move

# Example usage of the Tic-Tac-Toe functions
board = ['X', 'O', 'X', 'O', 'X', ' ', ' ', ' ', ' ']
print("Current Board:", board)

# Find the best decision for 'X'
move = best_decision(board)
print("Best move for X:", move)

# Apply the move and evaluate the new board
board[move] = 'X'
print("Board after move:", board)

score = evaluate(board)
if score == 1:
    print("Evaluated Score: X wins!")
elif score == -1:
    print("Evaluated Score: O wins!")
elif score == 0 and ' ' not in board:
    print("Evaluated Score: Draw!")
else:
    print("Evaluated Score: Game ongoing (or draw potential if no empty spaces left). Score:", score)
```
