```python
!pip install gTTS

from gtts import gTTS
from IPython.display import Audio, display
import re

print("Medical AI Assistant Ready...")

# Placeholder text as speech_recognition is not essential
text = "Patient John Doe has influenza. Prescribe Tamiflu and rest."
print("Simulated Recognized Text:", text)

try:
    # Simple medical information extraction
    disease = re.search(r"has ([^.]+?)\.", text) # Adjusted regex for diagnosis
    medicine = re.search(r"Prescribe (.+?)(?: and|,|\\.)", text)

    record = {
        "Diagnosis": disease.group(1).strip() if disease else "Not detected",
        "Prescription": medicine.group(1).strip() if medicine else "Not detected"
    }

    print("Structured Record:", record)

    confirmation = f"Diagnosis recorded as {record['Diagnosis']} and prescription as {record['Prescription']}."
    print("Confirmation:", confirmation)

    # Generate and play audio using gTTS
    tts = gTTS(text=confirmation, lang='en')
    tts.save("confirmation.mp3")
    display(Audio("confirmation.mp3", autoplay=True))

except Exception as e:
    print("Error:", e)
```
