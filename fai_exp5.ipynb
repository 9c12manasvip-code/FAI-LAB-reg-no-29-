```python
from collections import deque

# Matrix (1 = free, 0 = blocked)
grid = [[1,1,0],
        [1,1,1],
        [0,1,1]]

n, m = 3, 3

# ---- SORT MATRIX ----
flat = sorted([cell for row in grid for cell in row])
sorted_mat = [flat[i*m:(i+1)*m] for i in range(n)]

print("Sorted Matrix:")
for r in sorted_mat:
    print(r)

# ---- SHORTEST PATH (BFS) ----
dx, dy = [-1,1,0,0], [0,0,-1,1]
q = deque([(0,0,0)])  # x, y, distance
visited = set([(0,0)])

while q:
    x, y, d = q.popleft()
    if (x, y) == (2, 2):
        print("\nShortest Path Length:", d)
        break
    for i in range(4):
        nx, ny = x+dx[i], y+dy[i]
        if 0<=nx<n and 0<=ny<m and grid[nx][ny]==1 and (nx,ny) not in visited:
            visited.add((nx,ny))
            q.append((nx,ny,d+1))
```
