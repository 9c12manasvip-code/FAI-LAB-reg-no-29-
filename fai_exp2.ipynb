```python
import random

def score(S, cost, gain, limit):
    total_cost = sum(c*s for c,s in zip(cost, S))
    total_gain = sum(g*s for g,s in zip(gain, S))
    return total_gain if total_cost <= limit else 0

def hill_climb(cost, gain, limit, steps=100):
    n = len(cost)
    S = [random.randint(0,1) for _ in range(n)]

    for _ in range(steps):
        best = S
        for i in range(n):
            N = S[:]
            N[i] ^= 1
            if score(N, cost, gain, limit) > score(best, cost, gain, limit):
                best = N
        S = best
    return S, score(S, cost, gain, limit)

cost  = [2, 3, 4, 5]     # battery usage
gain  = [6, 8, 9, 12]    # information value
limit = 7               # battery limit

print(hill_climb(cost, gain, limit))
```
